```python
path = '/Users/koray/Desktop/misteli/min_dis/'

colszero = 0
it = 0
for i in gen:
    it += 1
    if it % 100 ==0:
        print(it, colszero)
    acc = i['accession']
    d1 = i['plate_id']
    d2 = i['plate_well_id']
    rc = d2.split('_')
    r = int(rc[1])
    c = int(rc[3])
    df_exp = df.loc[(df['assay.id'] == trans[d1]) & (df['nrow']==r) & (df['ncol'] == c)]
    cols = df_exp.shape[0]
    if cols == 0:
        #print(it, d1,d2)
        colszero += 1
        continue
    
    filename = path + acc + '.txt'
    df_exp.to_csv(filename)
    
print(colszero)
    

```
