```python
import glob
path = '/Users/koray/Desktop/Wrangling/Labs/Gianluca/4DN_Submission/'
files = []
files1 = glob.iglob(path + '**/*.xlsx', recursive=True)
files2 = glob.iglob(path + '**/*.xls', recursive=True)
for f in files1:
    files.append(f)
for f in files2:
    files.append(f)
print(len(f))
```

```python
allm = []
import xlrd
import pandas as pd
for f in files:
    tag = f.replace(path, "")
    try:
        bookread = xlrd.open_workbook(f)
    except:
        #print('CANT OPEN', tag)
        continue
        
    shs = bookread.sheet_names()
    if 'ExperimentMic' not in shs:
        #print("NO EXP MIC", tag)
        continue

    active_sheet = bookread.sheet_by_name('ExperimentMic')
    first_row_values = active_sheet.row_values(rowx=0)
    if '#plate_row' in first_row_values:
        if '#plate_column' in first_row_values:
            pass
            #print('USEFUL FILE', tag)
  
    df = pd.read_excel(f, 'ExperimentMic')
    df = df[~df['#Field Name:'].str.startswith("#", na = False)]
    for index, row in df.iterrows():
        allm.append([row['aliases'], row['plate_id'], row['#plate_row'],row['#plate_column']])
print(len(allm))
    

        
```
