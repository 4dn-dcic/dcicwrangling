```python
from dcicutils import ff_utils
from functions.notebook_functions import *

# get key from file
my_key = get_key('koray_data')
my_env = 'data'
# all usable env names
all_envs = ['data', 'staging', 'fourfront-webdev', 'fourfront-mastertest', 'fourfront-hotseat']
# get admin key from s3
#my_key = ff_utils.get_authentication_with_server({}, ff_env = 'data')

```

```python
users = ff_utils.search_metadata('/search/?type=user', key=my_key)
labs = ff_utils.search_metadata('/search/?type=lab', key=my_key)
awards = ff_utils.search_metadata('/search/?type=award', key=my_key)
```

```python

for user in users:
    if not user.get('lab'):
        #print('user with no lab', user['display_title'])
        continue
    lab = user['lab']
    if not lab.get('awards'):
        #print('user with no award', lab['display_title'])
        continue
    lab_awards = [i['uuid'] for i in lab['awards']]
    vgs = []
    for an_award in lab_awards:
        award = [i for i in awards if i['uuid'] == an_award]
        assert len(award) == 1
        vgs.append(award[0]['viewing_group'])
    vgs = list(set(vgs))
     
    vg = user.get('viewing_groups', [])
    
    patch = False
    for check_vg in vgs:
        if check_vg not in vg:
            if check_vg == '4DN':
                patch = True
                print('missing', check_vg)
                vg.append(check_vg)
            if check_vg == 'NOFIC':
                patch = True
                print('missing', check_vg)
                vg.append(check_vg)
    patch = False
    if patch:
        print('patching')
        ff_utils.patch_metadata({'viewing_groups':vg}, user['uuid'], key=my_key)
    
```
