```python
#get json file from s3
from dcicutils import ff_utils
from functions.notebook_functions import *
import json

my_key = get_key('koray_data')
q = '/search/?type%21=OntologyTerm&type=Item&field=uuid'
all_item_uuids = ff_utils.search_metadata(q, key=my_key)
print(len(all_item_uuids))
```

```python
import time
problem_report = []
counter = 0
start = time.time()
for an_item in all_item_uuids:
    counter += 1
    
    if counter % 100 == 0:
        print(counter, int((time.time()-start)/60), 'min')
        
    try:
        ff_utils.patch_metadata({}, obj_id = an_item['uuid'], key= my_key, add_on="check_only=True")
    except Exception as e:
        problem_report.append({'id':an_item['uuid'],
                               'error': e})
        print(problem_report)
        break
        
```
