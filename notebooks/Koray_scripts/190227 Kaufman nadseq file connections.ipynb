```python
# collect and clear files from exps
exps = '''kaufman:F121-9.replicate1.XL.genomic
kaufman:F121-9.replicate1.XL.nucleolar
kaufman:F121-9.replicate2.XL.genomic
kaufman:F121-9.replicate2.XL.nucleolar'''


for exp in exps.split('\n'):
    res = ff_utils.get_metadata(exp, my_key, add_on='frame=raw')
    print(res['accession'])
    files = res.get('files', [])
    if not files:
        print('no files')
    print(len(files))


        
# fs = '''kaufman:F121-9.replicate1.XL.genomic.R1_2
# kaufman:F121-9.replicate1.XL.genomic.R2_2
# kaufman:F121-9.replicate1.XL.genomic.R1_5
# kaufman:F121-9.replicate1.XL.genomic.R2_5
# kaufman:F121-9.replicate1.XL.nucleolar.R1
# kaufman:F121-9.replicate1.XL.nucleolar.R2
# kaufman:F121-9.replicate2.XL.genomic.R1_5_1
# kaufman:F121-9.replicate2.XL.genomic.R2_5_1
# kaufman:F121-9.replicate2.XL.genomic.R1_5_2
# kaufman:F121-9.replicate2.XL.genomic.R2_5_2
# kaufman:F121-9.replicate2.XL.nucleolar.R1_4
# kaufman:F121-9.replicate2.XL.nucleolar.R2_4
# kaufman:F121-9.replicate2.XL.nucleolar.R1_5
# kaufman:F121-9.replicate2.XL.nucleolar.R2_5'''
# fss = fs.split('\n')


# upl = ff_utils.search_metadata('/search/?status=uploading&track_and_facet_info.lab_name=No+value&type=FileFastq', my_key)
# upla = [i['aliases'][0] for i in upl]

# lf = [i for i in upla if i not in fss]
# print(lf)
# lfr = [i for i in upla if i in fss]
# print(lfr)


def fetch_pf_associated(pf_id, my_key):
    """Given a file accession, find all related items
    1) QCs
    2) wfr producing the file, and other outputs from the same wfr
    3) wfrs this file went as input, and all files/wfrs/qcs around it
    The returned list might contain duplicates, uuids and display titles for qcs"""
    file_as_list = []
    pf_info = ff_utils.get_metadata(pf_id, my_key)
    file_as_list.append(pf_info['uuid'])
    if pf_info.get('quality_metric'):
        file_as_list.append(pf_info['quality_metric']['uuid'])
    inp_wfrs = pf_info.get('workflow_run_inputs')
    out_wfrs = pf_info.get('workflow_run_outputs')
    for inp_wfr in inp_wfrs:
        file_as_list.extend(fetch_wfr_associated(inp_wfr['uuid'], my_key))
    for out_wfr in out_wfrs:
        file_as_list.extend(fetch_wfr_associated(out_wfr['uuid'], my_key))
    return list(set(file_as_list))
        
                
def fetch_wfr_associated(wfr_uuid, my_key):
    """Given wfr_uuid, find associated output files and qcs"""
    wfr_as_list = []
    wfr_info = ff_utils.get_metadata(wfr_uuid, my_key)
    wfr_as_list.append(wfr_info['uuid'])
    if wfr_info.get('output_files'):
        for o in wfr_info['output_files']:
                if o.get('value'):
                    wfr_as_list.append(o['value']['uuid'])
                elif o.get('value_qc'):
                    wfr_as_list.append(o['value_qc']) # this is a @id
    if wfr_info.get('output_quality_metrics'):
        for qc in wfr_info['output_quality_metrics']:
            if qc.get('value'):
                wfr_as_list.append(qc['value']['uuid'])
    return wfr_as_list

# for i in lf:
#     ff_utils.patch_metadata({'status':'deleted'}, i , my_key)
  
```
