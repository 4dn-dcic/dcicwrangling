```python
f2 = '/Users/koray/Desktop/misteli/submitted_target.txt'

probes = open(f2).read().split('\n')
probes = [i.split('\t') for i in probes]
print(len(probes))


file = '/Users/koray/Desktop/misteli/minDist.All.DCIC.txt'
import pandas as pd
df = pd.read_csv(file, sep = '\t')
df.shape
```

```python

def find_loc(chr, no, probes):
    chr = str(chr)
    no = str(no)
    if '.0' in no:
        no = no.split(".0")[0]
        
    found = False
    
    for ix in probes:
        if ix[1] == chr:
            if ix[4] == no:
                if ix[2] == 'missing':
                    print('need to fix')
                return [ix[0], ix[1], ix[2], ix[3]]

    print('not found', chr, no)
    return False
        
                
for i, row in df.iterrows():
    p_chr = row['chr']
    bait = row['Bait']
    target = row['Target']
    b_l = find_loc(p_chr, bait, probes)
    t_l = find_loc(p_chr, target, probes)
    
    if b_l:   
        df.at[i,'Bait Chr'] = b_l[1]
        df.at[i,'Bait Start'] = b_l[2]
        df.at[i,'Bait End'] = b_l[3]
    else:
        df.at[i,'Bait Chr'] = None
        df.at[i,'Bait Start'] = None
        df.at[i,'Bait End'] = None
        
    if t_l:
        df.at[i,'Target Chr'] = t_l[1]
        df.at[i,'Target Start'] = t_l[2]
        df.at[i,'Target End'] = t_l[3]
    else:
        df.at[i,'Target Chr'] = None
        df.at[i,'Target Start'] = None
        df.at[i,'Target End'] = None


df.shape
    
    
```
