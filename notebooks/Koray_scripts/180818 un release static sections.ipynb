```python
from dcicutils import ff_utils
from functions.notebook_functions import get_key

# get key from file
my_key = get_key('koray_data')

# all usable env names
all_envs = ['data', 'staging', 'fourfront-webdev', 'fourfront-mastertest', 'fourfront-hotseat']
# get admin key from s3
#my_key = ff_utils.get_authentication_with_server({}, ff_env = 'data')
```

```python
q_rel = '/search/?status=released&type=Page'
rel_pages = ff_utils.search_metadata(q_rel, key=my_key)

rel_sections = []
for a_page in rel_pages:
    print('PAGE', a_page['display_title'])
    if a_page.get('content'):
        for content in a_page['content']:
            rel_sections.append(content['uuid'])
            print(content['status'], content['display_title'])
    print()
print(len(rel_sections)) 
rel_sections = list(set(rel_sections))
print(len(rel_sections)) 
```

```python
q_drf = '/search/?status=draft&type=Page'
drf_pages = ff_utils.search_metadata(q_drf, key=my_key)

drf_sections = []
for a_page in drf_pages:
    print('PAGE', a_page['display_title'])
    if a_page.get('content'):
        for content in a_page['content']:
            drf_sections.append(content['uuid'])
            print(content['status'], content['display_title'])
    print()
print(len(drf_sections)) 
drf_sections = list(set(drf_sections))
print(len(drf_sections)) 
```

```python
for a_content in drf_sections:
    print(a_content)
    met = ff_utils.get_metadata(a_content, key=my_key)
    if met['status'] == 'released':
        ff_utils.patch_metadata({'status':'draft', 'uuid':a_content}, key = my_key)
        print('drafted')
```
