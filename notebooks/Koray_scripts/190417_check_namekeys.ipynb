```python
from dcicutils import ff_utils
from functions.notebook_functions import *
import re

my_key = get_key('koray_data')

['_', '-', ':', ',', '.', ' ', '@']

def validate(un_key):
    res = re.findall(r"[^a-zA-Z0-9\-:,.@_ ]", un_key)
    if res:
        msg = un_key + ' ' + str(res)
        return msg
    else:
        return
    
    
profiles = ff_utils.get_metadata('/profiles/', my_key, add_on='?frame=raw')
# run step1 - if item does not exist, post with minimal metadata
for a_type in profiles:
    if a_type in ['Biosample', 'ExperimentHiC', 'ExperimentSeq']:
        
        # this conversion of schema name to object type works for all existing schemas at the moment
        schema_info = profiles[a_type]
        ids = schema_info.get('identifyingProperties', [])
        if 'aliases' not in ids:
            ids.append('aliases')
        # print(a_type, ids)
        q = '/search/?type=' + a_type + '&field=' + '&field='.join(ids) + '&frame=raw'
        all_items = ff_utils.search_metadata(q, my_key)
        print(len(all_items))
    
        for an_item in all_items:
            for a_key in ids:
                if a_key in ['uuid']:
                    continue
                x = False
                a_case = an_item.get(a_key)
                if not a_case:
                    continue
                if isinstance(a_case, list):
                    for an_id in a_case:
                        x = validate(an_id)
                else:
                    x = validate(a_case)
                if x:
                    print(a_type, an_item['uuid'], a_key, end=" ")
                    print(x)
        print()


```
