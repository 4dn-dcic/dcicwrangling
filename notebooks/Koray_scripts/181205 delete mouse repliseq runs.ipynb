```python
from dcicutils import ff_utils
from functions.notebook_functions import *

# get key from file
my_key = get_key('koray_data')
my_env = 'data'
# all usable env names
all_envs = ['data', 'staging', 'fourfront-webdev', 'fourfront-mastertest', 'fourfront-hotseat']
# get admin key from s3
#my_key = ff_utils.get_authentication_with_server({}, ff_env = 'data')

```

```python
q = '/search/?experiments_in_set.biosample.biosource.individual.organism.name=mouse&experiments_in_set.experiment_type=Repli-seq&type=ExperimentSetReplicate'
sets = ff_utils.search_metadata(q, key = my_key)
print(len(sets))
```

```python
def cleaner(resp):
    pc = False
    opc = False
    if resp.get('processed_files'):
        pc = True
    if resp.get('other_processed_files'):
        opc = True
    if pc or opc:
        print('cleaning')
        pass
        ff_utils.patch_metadata({}, obj_id=resp['uuid'], key=my_key, add_on = 'delete_fields=processed_files,other_processed_files')

for i in sets: 
    cleaner(i)
    exps = i['experiments_in_set']
    for exp in exps:
        e_resp = ff_utils.get_metadata(exp['uuid'], key=my_key)                      
        cleaner(e_resp)
        files = e_resp.get('files')
        if not files:
            print('no file in exp', e_resp['accession'])
            continue
        for f in files:
            f_resp = ff_utils.get_metadata(f['uuid'], key=my_key)
            runs = f_resp['workflow_run_inputs']
            rep_runs = [r for r in runs if r['display_title'].startswith('repli')]
            print(len(rep_runs), 'many runs')
            
            for rep in rep_runs:
                patch_data = {'description': "This workflow run is deleted", 'status': "deleted"}
                ff_utils.patch_metadata(patch_data, obj_id=rep['uuid'] ,key=my_key)
                # delete output files of the deleted workflow run
                for o in rep['output_files']:
                    if o.get('value'):
                        ud = o['value']['uuid']
                        ff_utils.patch_metadata({'status': "deleted"}, obj_id=ud ,key=my_key)
                    
```

```python
#patch files to trigger indexing
it = 0
for i in sets: 
    it += 1
    print(it)
    exps = i['experiments_in_set']
    for exp in exps:
        e_resp = ff_utils.get_metadata(exp['uuid'], key=my_key)                      
        files = e_resp.get('files')
        if not files:
            print('no file in exp', e_resp['accession'])
            continue
        for f in files:
            ff_utils.patch_metadata({'notes': 'files'}, obj_id=f['uuid'], key=my_key)
           
```
