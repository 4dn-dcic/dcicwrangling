```python
# list of all scheams with an existing url field
profiles = ff_utils.get_metadata('/profiles/', key=my_key)
urls = []

for a_p in profiles:
    props = profiles[a_p]['properties']
    if 'url' in props.keys():
        urls.append(a_p)
    else:
        pass

for i in sorted(urls):
    print(i)
```

```python
my_env = 'data'
my_key = ff_utils.get_authentication_with_server({}, ff_env = my_env)
q = '/search/?references.uuid%21=No+value&type=Item'
with_ref = ff_utils.search_metadata(q, key=my_key)
print(len(with_ref))

all_types = []
counter=0

pubs = {}

for i in with_ref:
    counter += 1
    ref = i['references'][0]
    typ = i['@type'][0]
    
    if ref['uuid'] not in pubs:
        ref_data = ff_utils.get_metadata(ref['uuid'], key =my_key)
        pubs[ref['uuid']] = ref_data
    else:
        ref_data = pubs[ref['uuid']]
        
    refed = 'lonePub'
    if ref_data.get('exp_sets_prod_in_pub'):
        refed = 'HAS EXPS'

    
    if typ.startswith('Experiment') or typ in ['Biosample']:
        if refed == 'HAS EXPS':
            refed += ' deleted!'
            
            
    all_types.append(typ+' '+refed+' '+i['@id'])

for i in sorted(all_types):
    print(i)

    
```
