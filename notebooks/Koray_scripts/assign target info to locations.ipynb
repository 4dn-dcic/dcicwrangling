```python

def find_loc(chr, no, probes):
    chr = str(chr)
    no = str(no)
    if '.0' in no:
        no = no.split(".0")[0]
        
    found = False
    
    for ix in probes:
        if ix[1] == chr:
            if ix[4] == no:
                if ix[2] == 'missing':
                    print('need to fix')
                return [ix[0], ix[1], ix[2], ix[3]]

    print('not found', chr, no)
    return False
        
                
for i, row in df.iterrows():
    p_chr = row['chr']
    probe_green = row['Probe_Green'] 
    probe_red = row['Probe_Red']
    probe_farred = row['Probe_FarRed']
    
    g_l = find_loc(p_chr, probe_green, probes)
    r_l = find_loc(p_chr, probe_red, probes)
    f_l = find_loc(p_chr, probe_farred, probes)
    
    if g_l:   
        df.at[i,'Green Chr'] = g_l[1]
        df.at[i,'Green Start'] = g_l[2]
        df.at[i,'Green End'] = g_l[3]
    else:
        df.at[i,'Green Chr'] = None
        df.at[i,'Green Start'] = None
        df.at[i,'Green End'] = None
        
    if r_l:
        df.at[i,'Red Chr'] = r_l[1]
        df.at[i,'Red Start'] = r_l[2]
        df.at[i,'Red End'] = r_l[3]
    else:
        df.at[i,'Red Chr'] = None
        df.at[i,'Red Start'] = None
        df.at[i,'Red End'] = None
    
    if f_l:
        df.at[i,'FarRed Chr'] = f_l[1]
        df.at[i,'FarRed Start'] = f_l[2]
        df.at[i,'FarRed End'] = f_l[3]
    else:
        df.at[i,'FarRed Chr'] = None
        df.at[i,'FarRed Start'] = None
        df.at[i,'FarRed End'] = None


df.shape
    
    
```
