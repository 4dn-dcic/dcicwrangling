### For a set of replicates - see if any linked items have GEO dbxrefs

```python
from dcicutils import ff_utils
from functions.notebook_functions import *
from functions.wfr import *

my_auth = get_key('andyprod', keyfile='~/keypairs.json')

sets_in_scope = []

search_url  = '/search/?type=ExperimentSetReplicate&status=released&experiments_in_set.experiment_type.experiment_category=Sequencing' # /search/?type=ExperimentSetReplicate&experimentset_type=replicate&lab.display_title=Sheng+Zhong%2C+UCSD&status=pre-release'

if sets_in_scope:
    esets = [ff_utils.get_metadata(i, my_auth) for i in sets_in_scope]
elif search_url:
    esets = [i for i in ff_utils.search_metadata(search_url, my_auth)]

def has_geo_dbxref(dbxrefs):
    return any([a for a in dbxrefs if a.startswith('GEO')])

def has_sra_dbxref(dbxrefs):
    return any([a for a in dbxrefs if a.startswith('SRA')])

    
in_geo = 0
in_sra = 0
not_in_geo = 0
total = 0

print("processing {} esets".format(len(esets)))
for eset in esets:
    total += 1
    es_acc = eset.get('accession')
    if has_geo_dbxref(eset.get('dbxrefs', [])):
        # print(f"{es_acc} in GEO")
        in_geo += 1
        continue
    exps = eset.get('experiments_in_set', [])
    found = False
    for exp in exps:
        euuid = exp.get('uuid')
        expmeta = ff_utils.get_metadata(euuid, my_auth)
        dbx = expmeta.get('SRxrefs', [])
        if has_geo_dbxref(dbx):
            # print(f"{es_acc} in GEO")
            in_geo += 1
            found = True
            break
        if has_sra_dbxref(dbx):
            in_sra += 1
            print(f"{es_acc} ONLY IN SRA")
            found = True
            break
    if not found:
        dsl = eset.get('dataset_label', 'N.A.')
        print(f"{es_acc} {dsl} NOT IN!")
        not_in_geo+= 1

print(f"IN GEO {in_geo}")
print(f"NOT {not_in_geo}")
print(in_geo/total)
print(f"IN SRA ONLY {in_sra}")
print(total)
    
            
    
```
